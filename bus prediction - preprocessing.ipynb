```python
#A summary of missing variables represented as a percentage of the total missing content.
def missingness_summary(df, print_log=False, sort='ascending'):
  s = df.isnull().sum()*100/df.isnull().count()

  if sort.lower() == 'ascending':
    s = s.sort_values(ascending=True)
  elif sort.lower() == 'descending':
    s = s.sort_values(ascending=False)
  if print_log:
    print(s)

  return pd.Series(s)
```

# At the Stop

Investigating situations where the buses are at the stop already. DistanceFromStop should equal zero.

Columns of interest
* RecordedAtTime
* PublishedLineName
* DestinationName
* VehicleLocation.Longitude
* VehicleLocation.Latitude

* VehicleRef

```python
#get the rows and columns of interest.
coi = ['RecordedAtTime','PublishedLineName','DirectionRef','VehicleRef','DestinationLat','DestinationLong', 'NextStopPointName','OriginLat','OriginLong','VehicleLocation.Latitude', 'VehicleLocation.Longitude','DistanceFromStop','ScheduledArrivalTime']
atthestop = transit[coi].loc[transit.DistanceFromStop == '0'].copy()
atthestop.info()
```

```python
#Convert the lat and long to numeric.
atthestop['VehicleLat'] = atthestop['VehicleLocation.Latitude'].apply(pd.to_numeric, errors='coerce')
atthestop['VehicleLong'] = atthestop['VehicleLocation.Longitude'].apply(pd.to_numeric, errors='coerce')
atthestop.drop(columns=['VehicleLocation.Latitude', 'VehicleLocation.Longitude'], inplace=True)
atthestop
```

```python
#Recorded at time converted to date time.
atthestop['RecordedAtTime'] = pd.to_datetime(atthestop['RecordedAtTime'], errors = 'coerce')
tindex = pd.DatetimeIndex(atthestop.RecordedAtTime)
#Recorded Year, Day, and Month as a separate column for convenience
atthestop['RecordDate'] = tindex.strftime('%Y-%m-%d ')
```

```python
#Troubleshooting scheduled arrival time.
sTimes = atthestop[['ScheduledArrivalTime', 'RecordDate']].loc[atthestop.ScheduledArrivalTime >= "24"]
```

```python
#year-month-day as string
ymd = sTimes.RecordDate
hours = sTimes.ScheduledArrivalTime.str[0:2]
rest = sTimes.ScheduledArrivalTime.str[2:]
# subtract 24
hours = hours.astype(int) - 24
# reassign the string
converted = hours.astype(str) + rest
converted = ymd + converted
# make it datetime, adding one to the day because it's after midnight.
converted = pd.to_datetime(converted, format = '%Y-%m-%d %H:%M:%S')
converted = converted
```

```python
from datetime import datetime, timedelta
#scheduled column as Scheduled arrival time converted to date time.
atthestop['scheduled'] = pd.to_datetime(atthestop['RecordDate'] + atthestop['ScheduledArrivalTime'], format = '%Y-%m-%d %H:%M:%S', errors='coerce')
atthestop['scheduled'].fillna(value=converted, inplace=True)

#correct for situations after midnight.
condition = (atthestop.scheduled - atthestop.RecordedAtTime).astype('timedelta64[s]') >= timedelta(minutes=1000)
atthestop.loc[condition, ['scheduled']] = atthestop.loc[condition, ['scheduled']] - timedelta(days=1)

condition = (atthestop.scheduled - atthestop.RecordedAtTime).astype('timedelta64[s]') <= timedelta(minutes=-1000)
atthestop.loc[condition, ['scheduled']] = atthestop.loc[condition, ['scheduled']] + timedelta(days=1)

#late indicator
condition = (atthestop.scheduled - atthestop.RecordedAtTime).astype('timedelta64[s]') < timedelta(minutes=0)
atthestop['late'] = condition

#adherence
atthestop['adherence'] = (atthestop.scheduled - atthestop.RecordedAtTime).astype('timedelta64[s]') / pd.Timedelta(minutes=1)
atthestop['adherence'] = atthestop['adherence'].astype('int')

#Calculate the day of the week
atthestop['day_of_week']= atthestop['scheduled'].dt.strftime("%A")
```

```python
weather = pd.read_csv('/kaggle/input/nyc-weather/New-York-CityUSA-2017-06-01-to-2017-06-30.csv', low_memory=False)

# Display the first few rows of the DataFrame
weather.head()
```

```python
weather['datetime'] = pd.to_datetime(weather['datetime'])
weather
```

```python
#Augment the vehicle data with weather information at that hour
def add_weather_data(data):
    for index, row in data.iterrows():
        wrow = weather.iloc[(weather['datetime'] - row['RecordedAtTime']).abs().argsort()[:1]]
        data.at[index,'Temp'] = list(wrow['temp'])[0]
        data.at[index,'Pressure'] = list(wrow['sealevelpressure'])[0]
        data.at[index,'Humidity'] = list(wrow['humidity'])[0]
        data.at[index,'WindSpeed'] = list(wrow['windspeed'])[0]
        data.at[index,'Clouds'] = list(wrow['cloudcover'])[0]
        data.at[index,'Visibility'] = list(wrow['visibility'])[0]
        data.at[index,'WeatherConditions'] = list(wrow['conditions'])[0]
    return(data)
```

```python
#Ok, seems right
data = add_weather_data(atthestop)
data
```
